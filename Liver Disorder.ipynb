```python
import pandas as pd
data=pd.read_csv('/content/Indian Liver Patient Dataset (ILPD).csv')
data
```

```python
target=data['Selector']
target.head()
```

```python
x=data.drop('Selector',axis=1)
x.head()
```

```python
x.isnull().sum()
```

```python
x.replace('Male',1, inplace=True)
x.replace('Female',0, inplace=True)
x.isnull().sum()
```

```python
x.head()
```

```python
x['A/G Ratio'].fillna(x['A/G Ratio'].mean(), inplace=True)
```

```python
x.isnull().sum()
```

```python
print(x)
```

```python
x.info()
```

```python
x.describe()
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x,target=smote.fit_resample(x,target)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,target,test_size=0.2,random_state=42)
```

```python
x_train
```

#Perceptron

```python
from sklearn.linear_model import Perceptron
pm=Perceptron()
pm.fit(x_train,y_train)
```

```python
yp=pm.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(y_test,yp))
```

```python
print(classification_report(y_test,yp))
```

```python
print(accuracy_score(y_test,yp))
```


#LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
yp = lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(y_test,yp))
```

```python
print(classification_report(y_test,yp))
```

```python
print(accuracy_score(y_test,yp))
```

#Support Vector Mechine

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train,y_train)
```

```python
yp=sv.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(y_test,yp))
```

```python
print(classification_report(y_test,yp))
```

```python
print(accuracy_score(y_test,yp))
```
